```python
#Reasons for leaving job using infromation from dataset (Reasons for termination)
df = pd.read_csv("HRDataset.csv")
df = df[df.TermReason != 'N/A-StillEmployed']
df['TermReason']
df['TermReason'].value_counts()
ax=df['TermReason'].value_counts().plot(kind='barh',figsize = (15,10), title= 'Reasons for leaving Job')
ax.set_ylabel("Reasons for Leaving")
ax.set_xlabel("Frequency")
# Most people leave their jobs for another position elsewhere and due to unhappiness.
```
