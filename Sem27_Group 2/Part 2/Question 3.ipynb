```python
#employment rate in general
import pandas as pd
import matplotlib.pyplot as plt
import datetime
df = pd.read_csv("employment_rate.csv")
date = pd.to_datetime(df[['year','month','day']])
df['Date'] = date
df.drop(labels=['Date'], axis=1,inplace = True)
df.insert(0, 'Date', date)
column_mean = df.loc[:,"emp_combined_ss40":'emp_combined_ss70']
df['employment_mean'] = column_mean.mean(axis=1)
df.plot.line(x='Date',y='employment_mean')
```

```python
#Importing Employment Rate CSV
employmentData = pd.read_csv('employment_rate.csv')
#Converting to date time format
employmentData['datetime'] = pd.to_datetime(employmentData[["year","month", "day"]])
#Plotting line graph
plt.figure(figsize=(10, 10))
#Trade, Transportation and Utilities Sector
plt.plot( 'datetime', 'emp_combined_ss40', data = employmentData, color='orange', linewidth=2)
#Professional and Business Services Sector
plt.plot( 'datetime', 'emp_combined_ss60', data = employmentData, color='magenta', \
linewidth=2)
#Education and Health Services Sector
plt.plot( 'datetime', 'emp_combined_ss65', data = employmentData, color='limegreen',\
linewidth=2)
#Leisure and Hospitality Sector
plt.plot( 'datetime', 'emp_combined_ss70', data = employmentData, color='red', linewidth=2)
plt.legend()
plt.title("Employment Rates in 2020 by Sector", size = 20)
plt.xlabel("Months", size = 16)
plt.ylabel("Employment Level", size = 16)
#general decline in employment in April and May most likely to due covid-19 pandamic and the lockdown. However, employment rate is steadily increasing since June.
```
