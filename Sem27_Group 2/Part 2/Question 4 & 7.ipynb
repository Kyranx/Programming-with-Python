```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("absentees.csv")
df.isnull().sum() #no null values
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df1=df.groupby(["Work load Average/day "]).mean()["Absenteeism time in hours"]
plot1=df1.reset_index().plot(kind="hist", x="Absenteeism time in hours", bins=10, title=" Average Absenteeism Based on Workload")
plot1.set_ylabel("Average Absenteeism Time")
plot1.legend(["Average Absenteeism"])
#people with lesser work tend to be absent more, likely becasue they have nothing or few things to do during work hours.
#qns 7
```

```python
import pandas as pd
from matplotlib import pyplot as plt
#Importing Absentees CSV
df4 = pd.read_csv("absentees.csv")
#Checking and dropping duplicate records in selected columns relevant to the question
df4.duplicated(["ID","Reason for absence","Month of absence","Day of the week","Absenteeism time in hours"]).sum()
df4= df4.drop_duplicates(["ID","Reason for absence","Month of absence","Day of the week",\
"Absenteeism time in hours"])
#Checking for missing values
df4.isna().sum()
#Dropping rows with zero hours of absence
df4 = df4.drop(df4[df4["Absenteeism time in hours"]==0].index)
#Finding out the main reason for absence
df4["Reason for absence"].mode()
#Number of occurrences for each reason and plotting bar graph
no_of_occurences = df4["Reason for absence"].value_counts()
chart = no_of_occurences.plot(kind='bar', figsize = (15,10), title = "Reasons for employees to be absent from work",\
color='lightseagreen')
chart.set_xlabel("\nReason for Absence")
chart.set_ylabel("Frequency")
#qns 4
```

```python
df = pd.read_csv('employee.csv')
df.dropna(inplace=True)
df['Age'].value_counts()
Young = 0
Middle = 0
Old = 0
for each in df['Age']:
    if each <25:
        Young += 1
    elif 25 <= each<45:
        Middle += 1
    elif each>=45:
        Old += 1
print(Young)
print(Middle)
print(Old)
test = [Young,Middle,Old]
coordinate = ['Below 25','Between 25 & 45','45 & above']
plt.bar(coordinate,test,linestyle='--')
plt.xlabel("Age")
plt.ylabel("Number of Employees")
#Employee age information (supporting data)
```
