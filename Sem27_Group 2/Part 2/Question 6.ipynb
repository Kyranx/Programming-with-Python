```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("glassdoor_salary_dataset.csv")
df.isnull().sum() #no null values
```

```python
df.duplicated().sum() #no duplicated
```

```python
df1=df.drop(["age", "perfEval", "seniority"], axis=1)
salary=df1.groupby("dept").mean()
salary.sort_values(by="basePay",ascending=True).plot(kind="bar", figsize=(20,10))
#qns 6
#Salary Trend based on department
```

```python
salary2=df1.groupby("jobTitle").mean()
salary2.sort_values(by="basePay",ascending=True).plot(kind="bar", figsize=(20,10))
#qns 6
#Salary trend based on job role
```

```python
salary3=df.groupby("perfEval").mean()["bonus"]
salary3.plot(kind="bar", figsize=(20,10))
salary3=plt.xlabel('Performance Evaluiation')
#qns 6
```

```python
#Importing glassdoor salary dataset Absence Rate
df2 = pd.read_csv('glassdoor_salary_dataset.csv') 
#Group by job title to view statistics of salaries
salary = df2.groupby(['jobTitle'])
pd.DataFrame(salary['basePay'].describe()) 
#Plotting horizontal bar graph
y = pd.DataFrame(salary['basePay'].describe())
y.drop(columns=['count', 'std', '75%', '25%', '50%'],inplace = True)
y.plot(kind = 'barh')
plt.rcParams['figure.figsize'] = (15,10)
plt.xlabel('Salary')
plt.ylabel('Job Title')
plt.title('Salary Statistics by Job Title')
```
